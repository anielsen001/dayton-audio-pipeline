Loading data
============

```python
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
%matplotlib inline
```

```python
raw = json.load(open('timings.json'))
```

```python
clean = [{'bytes': r['size_bytes'], 
          'kb': r['size_bytes'] / 1024, 
          'processing_seconds': r['elapsed_seconds'], 
          'file_seconds': r['file_stats']['seconds']} 
         for r in raw.values()]
```

```python
df = pd.DataFrame(clean)
```

File size vs. Watson time
=========================

```python
df.plot(x='kb', y='processing_seconds', kind='scatter')
```

Zoom on plot origin
===================

```python
df.plot(x='kb', y='processing_seconds', kind='scatter', xlim=[0, 2000], ylim=[0, 40])
```
